Predict flight delays based on features like departure time, carrier, and weather

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Features and target variable
features = df[['DepHour', 'CRSDepTime', 'WeatherDelay', 'CarrierDelay', 'LateAircraftDelay']]
target = df['ArrDelay']

# Handle missing values
features = features.fillna(features.mean())
target = target.fillna(target.mean())
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predictions
predictions = rf.predict(X_test)

# Evaluate accuracy of the model using mse
mse = mean_squared_error(y_test, predictions)
print('Mean Squared Error:',mse)
```

Inference:
 CarrierDelay and LateAircraftDelay have a significant impact on the model's predictions of arrival delays. These are critical variables that affect flight delays.

#Feature Engineering
Modify features based on insights from feature importance.

```python
# Creating a new feature: Interaction term between CarrierDelay and WeatherDelay
features['CarrierWeatherInteraction'] = features['CarrierDelay'] * features['WeatherDelay']
features['CarrierWeatherInteraction']
```

```python
# Feature transformation: Binning the departure hour into morning, afternoon, evening, and night
features['DepHourBin'] = pd.cut(features['DepHour'], bins=[0, 6, 12, 18, 24], labels=['Night', 'Morning', 'Afternoon', 'Evening'])
features['DepHourBin']
```

```python
# Convert categorical features to dummy variables
features = pd.get_dummies(features, columns=['DepHourBin'])
features
```

Alternative algorithm to improve model performance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Re-split the data with new features
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Fill missing values
features = features.fillna(features.mean())
target = target.fillna(target.mean())

# Train Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Make predictions
y_pred = lr.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error: ", mse)
print("R-squared: ",r2)
```

Get coefficients and feature names

```python
coefficients = lr.coef_
feature_names = X_train.columns
```

Create DataFrame for coefficients

```python
coefficients_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
}).sort_values(by='Coefficient', ascending=False)

# Display coefficients
print("Coefficients:\n", coefficients_df)

```

Plot coefficients

```python
plt.figure(figsize=(12, 8))
plt.barh(coefficients_df['Feature'], coefficients_df['Coefficient'], color='skyblue')
plt.xlabel('Coefficient Value')
plt.title('Feature Coefficients in Linear Regression Model')
plt.gca().invert_yaxis()
plt.show()
```

plt.gca().invert_yaxis():
Inverts the y-axis of the plot.
 By default, the highest values are plotted at the top. Inverting the y-axis ensures that features with higher coefficients appear at the top of the plot, making it easier to see which features have the most significant impact.

Inference:
Features with positive coefficients increase the predicted delay. CarrierDelay has a high positive coefficient, it means that as CarrierDelay increases, the model predicts a longer flight delay.
Features with negative coefficients decrease the predicted delay. DepHour has a negative coefficient, it implies that flights departing at certain hours are associated with shorter delays according to the model.
